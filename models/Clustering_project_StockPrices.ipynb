```python
from math import sqrt
returns = df.iloc[:,1:].pct_change().mean() * 252
variance = df.iloc[:,1:].pct_change().std() * sqrt(252)
```

```python
returns.columns = ["Returns"]
variance.columns = ["Variance"]
```

```python
ret_var = pd.concat([returns, variance], axis = 1).dropna()
ret_var.columns = ["Returns","Variance"]
```

```python
ret_var
```

```python
X =  ret_var.values
X
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', random_state = 60)
y_kmeans = kmeans.fit_predict(X)
print(y_kmeans)
```
